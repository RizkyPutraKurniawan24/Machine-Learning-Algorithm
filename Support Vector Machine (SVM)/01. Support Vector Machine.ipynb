# **Definition of Support Vector Machine (SVM)**

## **First Problems**

```python
demo.problem(1)
```

```python
illustration.choose_one
```

**Note**

Solution : Looking for the best and widest margin.

SVM was formerly known as the Maximum Margin Classifier. It turns out that an alternative margin is by finding support vectors (the circled data). That is why this algorithm is also called the Support Vector Machine (SVM), which means a machine that searches for support vectors.

```python
illustration.maximum_margin
```

**Note**

$C$ - Penalty Parameter

>- The larger $C$ == The greater the penalty for errors == more sensitive.
>- The smaller $C$ == The more tolerant of errors == smoother.

```python
demo.C(1)
```

**Note**

Solution : Big picture vs Detail Oriented == kernel Coefficient.

As $\gamma$ increases == more detail oriented == more sensitive.

As $\gamma$ decreased == more big picture oriented == smoother.

```python
demo.gamma()
```

```python
demo.svm()
```

**Note**

Avoid using a gamma that is too large, because it tends to overfit.

Tips : This can also be addressed using scaling. Scaling is also known to help SVM.
